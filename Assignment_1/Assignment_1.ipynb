```python
import numpy as np
import random 
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Set up the sample size function 
random.seed(624) 

def sample_size(n):
    '''This function is for apply the different size of n
    
    generate the random uniform data which between 0 and 1 for different size of n'''
    
    sample = np.random.uniform(0,1,(10000,n))
    
    return sample


n10 = sample_size(10) # n =10
n100 = sample_size(100) # n =100
n1000 = sample_size(1000) # n =1000
n10000 = sample_size(10000) # n =10000
```

```python
# uniform distribution function to determine the sample is 1 or -1
def uniform(n): 
    '''this is the uniform function for the samples we generate from previous function.
    
    set the result to be a list.
    if x less than 0.5, we append the -1 to the result
    if x greater than 0.5, we append 1 to the result.
    then return the average of the result'''

    result = []
    
    for x in n:
        if x < 0.5:
            result.append(-1)
        else:
            result.append(1)
            
    return np.mean(result)

```

```python
# use map() to returns a list of the results after applying the uniform function to each item of different n
average_n10 = list(map(uniform,n10)) 
average_n100 = list(map(uniform,n100))
average_n1000 = list(map(uniform,n1000))
average_n10000 = list(map(uniform,n10000))
```

# (a) log<sub>10</sub>(n) v.s. $ \bar{X}_n^{(1)} $ - $ \mu $

```python
#log10(n) vs. Xn(1)

n= [10,100,1000,10000] 

#log10 of different n
ln = np.log10(n) 

#get the first dataset in each average of different dataset
xbar = [average_n10[0],average_n100[0],average_n1000[0],average_n10000[0]] 

# plot
plt.plot(ln,xbar)
plt.ylabel('Average of n') #set y label
plt.xlabel('Number of ln') #set x label
plt.title(" log10(n) VS. Average of n")

plt.show()
```

# (b) Draw log<sub>10</sub>(n) v.s. $ \frac{1}{N} \sum_{i=1}^N \amalg $ { $ \left | $ $\bar{X}_n^{(1)} $ - $ \mu $ $ \right | $ $ > \varepsilon $}


```python
def xbarn(epsilon,average_n):
    '''This function is for the sum of the xbar in different n
    
    when each dataset in average of n greater than 0.5, count it as 1
    other we count as 0
    after sum toghter / lengh of the total average number of dataset we will get the result for different n'''
    
    ans = []
    mu = 0
    for i in average_n:
        if abs(i-mu) > epsilon:
            ans.append(1)
        else:
            ans.append(0)

    return sum(ans)/len(average_n)

# epsilon = 0.5
totale_n10 = xbarn(0.5,average_n10)
totale_n100 = xbarn(0.5,average_n100)
totale_n1000 = xbarn(0.5,average_n1000)
totale_n10000 = xbarn(0.5,average_n10000)

# epsilon = 0.1
totalp_n10 = xbarn(0.1,average_n10)
totalp_n100 = xbarn(0.1,average_n100)
totalp_n1000 = xbarn(0.1,average_n1000)
totalp_n10000 = xbarn(0.1,average_n10000)

# epsilon = 0.05
totals_n10 = xbarn(0.05,average_n10)
totals_n100 = xbarn(0.05,average_n100)
totals_n1000 = xbarn(0.05,average_n1000)
totals_n10000 = xbarn(0.05,average_n10000)
```

## Convert each image into a vector and form a matrix, which is of size number of images * number of dimensions and standardize it.

```python
import numpy as np
from numpy import *
### cv2 reads images
### glob operates Unix style pathname pattern expansion
import cv2
import glob
### importing functools for reduce() 
import functools
### improting preprocessing model for standandaze the data
from sklearn import preprocessing
### importing PCA for process the pca buildin function
from sklearn.decomposition import PCA
### improting the pandas for transform the pc coord to the dataframe
import pandas as pd
```

```python
### function Readfile
### input: a given pattern with path and form for glob function
### output: a combined matrix with info from the files
def Readfile(pattern):
    filenames = glob.glob(pattern)
    filenames.sort()
    ## read the files in pattern
    ims = [cv2.imread(im, -1).flatten() for im in filenames]
    ## return a combined matrix using reduce method
    return functools.reduce( lambda a,b : np.vstack((a,b)),ims)
### the patterns
patterns = ["../Assignment 1/homework1-data/image0/*.pgm",
            "../Assignment 1/homework1-data/image1/*.pgm",
            "../Assignment 1/homework1-data/image2/*.pgm"]
```

```python
### Mat is the final matrix we are dealing with, reduce again
Mat = functools.reduce( lambda a,b : np.vstack((a,b)),list(map(Readfile,patterns))) 
Mat.shape
```

```python
#standarized 
MatS = (Mat - mean(Mat)) / std(Mat)
MatS

#reference: https://stackoverflow.com/questions/4544292/how-do-i-standardize-a-matrix
```

## Perform PCA on the standardized data matrix, with number of components= 2, using Python's in-built command.

```python
pca = PCA(n_components=2)
pca
```

## Transform the data into the two PC coordinates and draw a scatter-plot of principal component 1 versus principal component 2.

```python
principalComponents = pca.fit_transform(MatS) ## transforming the data into PC coordinates.
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2']) # transfrom the pc coord to the data frame
pc1 = principalDf.iloc[:,0] #label pc1
pc2 = principalDf.iloc[:,1] #label pc2
#principalDf[0:10]
pc1.shape #check the pc1 shape
pc2.shape
# since we have 3 type of data, and each of them are 50000 so we have 15000 data
```

```python
# plot the data 
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

ax.scatter(pc1[0:5000],pc2[0:5000], c = 'red', s = 10, label = 'image0')
ax.scatter(pc1[5000:9999],pc2[5000:9999], c = 'blue', s = 10,label ='image1')
ax.scatter(pc1[9999:14999],pc2[9999:14999], c = 'green', s = 10,label ='image2')
ax.legend()

ax.grid()
```
